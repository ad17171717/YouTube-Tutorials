# **OpenCV's `resize` Function and Image Transformations**

**OpenCV's `resize` function can be used to enlarge or reduce an image's size.**


**The package provides different algorithms to perform geometrical transformations of a given image. The functions deform the pixel grid and map the grid to the destination image. The mapping is done in the reverse order, from destination to the source.**

| **Python Function**  | **Algorithm Description**  |
|---|---|
|**cv.INTER_NEAREST**   | **Nearest neighbor interpolation**  |
|**cv.INTER_LINEAR**   | **Bilinear interpolation**  |
| **cv.INTER_CUBIC**  | **Bicubic interpolation**  |
| **cv.INTER_AREA**  | **Resampling using pixel area relation. It may be a preferred <br />method for image decimation, as it gives moire'-free results.<br /> But when the image is zoomed, it is similar to the INTER_NEAREST method.**   |
| **cv.INTER_LANCZOS4**  | **Lanczos interpolation over 8x8 neighborhood**  |
| **cv.INTER_LINEAR_EXACT**  | **Bit exact bilinear interpolation**  |
| **cv.INTER_NEAREST_EXACT**  | **Bit exact nearest neighbor interpolation. This will produce<br /> same results as the nearest neighbor method in PIL, scikit-image or Matlab.**  |
| **cv.INTER_MAX**  | **Mask for interpolation codes**   |
| **cv.WARP_FILL_OUTLIERS**  | **flag, fills all of the destination image pixels. If some of them correspond to<br /> outliers in the source image, they are set to zero**  |
| **cv.WARP_INVERSE_MAP**  | **Flag, inverse transformation<br />For example, linearPolar or logPolar transforms:<br /><br />-flag is not set: $dst(,)=src(x,y)$<br /><br />-flag is set: $dst(x,y)=src(,)$**  |

<sup>Source: [Geometric Image Transformations](https://docs.opencv.org/3.4/da/d54/group__imgproc__transform.html) from OpenCV's Documentation<sup/>

## **Enlarging an Image's Size**

```python
enlg_img_1 = cv.resize(img, dsize=(1100,1100),interpolation=cv.INTER_CUBIC)
cv2_imshow(enlg_img_1)
```

```python
enlg_img_2 = cv.resize(img, dsize=(1100,1100),interpolation=cv.INTER_NEAREST)
cv2_imshow(enlg_img_2)
```

```python
enlg_img_3 = cv.resize(img, dsize=(1100, 1100), interpolation=cv.INTER_AREA)
cv2_imshow(enlg_img_3)
```

```python
enlg_img_1.shape
```

```python
np.allclose(enlg_img_1,enlg_img_2,enlg_img_3)
```

## **Reducing an Image's Size**

```python
red_img_1 = cv.resize(img,dsize=(200,200),interpolation=cv.INTER_CUBIC)
cv2_imshow(red_img_1)
```

```python
red_img_2 = cv.resize(img, dsize=(200, 200), interpolation=cv.INTER_NEAREST)
cv2_imshow(red_img_2)
```

```python
red_img_3 = cv.resize(img, dsize=(200, 200), interpolation=cv.INTER_AREA)
cv2_imshow(red_img_3)
```

```python
red_img_1.shape
```

```python
np.allclose(red_img_1,red_img_2,red_img_3)
```
