# **Polynomial**

**A polynomial is a mathematical expression involving a sum of powers in one or more variables multiplied by coefficients.**

**A polynomial can have constants, variables and exponents but never division by a variable.**

<sup>Source: [Polynomial](https://mathworld.wolfram.com/Polynomial.html) by Eric W. Weisstein from Wolfram MathWorld<sup/>
    
<sup>Source: [Polynomials](https://www.mathsisfun.com/algebra/polynomials.html) from MathIsFun.com<sup/>

## **Quadratic Function**

**A quadratic function is a polynomial function with one or more variables in which the highest-degree term is of the second degree.**

**A quadratic function is a function of the form: $f(x) = ax^2 + bx +c, \ \ \ a \neq 0$**

<sup>Source: [Quadratic Function](https://en.wikipedia.org/wiki/Quadratic_function) from Wikipedia.org<sup/>
    
<sup>Source: [College Algebra 7th Edition](https://www.cengage.com/c/college-algebra-7e-stewart/9781305115545/) by James Stewart, Lothar Redlin and Saleem Watson<sup/>

## **Example 1**

**Evaluate the function below when $x = 5$.**

$x^2 + 3x + 2$

```python
def poly_func(x):
    return x**2 + 3*x + 2
```

```python
poly_value = poly_func(5)

print(f'When x = 2 the polynomial returns {poly_value}')
```

```python
print(f'When x = 5 the polynomial returns {poly_value}')
```

## **Example 2**

**Graph the polynomial equation: $x^2 + 3x + 2$ where the range for the x-values is $[-5,5]$.**

```python
num_array = np.arange(-5,5,.01)
poly_values = [poly_func(x) for x in num_array]
```

```python
plt.plot(num_array,poly_values)
plt.axvline(0,color='black')
plt.axhline(0,color='black')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid();
```

## **Example 3**

**Find the vertex (maximum or minimum value) of the polynomial equation: $x^2 + 3x + 2$.**

**The maximum or minimum value of a quadratic function $f(x) = ax^2 + bx + c$ occurs at $x = - \large{\frac{b}{2a}}$**

- **If $a > 0$ then the minimum value is $\large{f(-\frac{b}{2a})}.$**
- **If $a < 0$ then the maximum value is $\large{f(-\frac{b}{2a})}.$**

<sup>Source: [College Algebra 7th Edition](https://www.cengage.com/c/college-algebra-7e-stewart/9781305115545/) by James Stewart, Lothar Redlin and Saleem Watson<sup/>

<center>$\Huge{x = - \frac{3}{(2)(1)}}$<center/>

<center>$\Huge{x = - \frac{3}{2}}$<center/>

<center>$\Huge{-(\frac{3}{2})^2 + 3(-\frac{3}{2})+2} = -\frac{1}{4} \text{ is the global minimum}$<center/>

```python
poly_func(-(3/2)) == min(poly_values)
```

```python
num_list = list(num_array)

plt.plot(num_array,poly_values)
plt.scatter(num_list[poly_values.index(min(poly_values))],min(poly_values),color='red',s=100)
plt.axvline(0,color='black')
plt.axhline(0,color='black')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid();
```

# **Application: Measuring the Gas Mileage of a Car**

**Most cars get their best gas mileage when traveling at a relatively modest speed. The gas mileage $M$ for a certain new car is modeled by the function: $M(s) = -\frac{1}{28}s^2+3s-31 \text{, when } 15 \leq s \leq 70$**

<sup>Source: [College Algebra 7th Edition](https://www.cengage.com/c/college-algebra-7e-stewart/9781305115545/) by James Stewart, Lothar Redlin and Saleem Watson<sup/>

## **Question 1**

**What is the average gas mileage if you drive 50 miles per hour for 1 hour and 65 miles per hour for 30 minutes?**

```python
def mileage_calc(speed):
    return -(1/28) * speed**2 + 3*speed -31
```

```python
avg_gas_mileage = mileage_calc(50) * (1/1.5) + mileage_calc(65) * (.5/1.5)

print(f'The average gas mileage for the trip is {avg_gas_mileage:.2f}')
```

## **Question 2**

**At what speed does the car consume the least gas per mile (highest gas mileage)?**

<center>$\Large{\text{Optimal Speed} = - \frac{3}{(2)(-\frac{1}{28})} = 42 \text{ miles per hour}}$<center/>

<center>$\Large{\text{Optimal Gas Mileage} = -\frac{1}{28}(42)^2+3(42)-31} = 32 \text{ miles per gallon of gas}$<center/>

```python
speed_list = np.arange(15,70,.01)
mileage_list = [mileage_calc(speed) for speed in speed_list]
```

```python
mileage_calc(42) == max(mileage_list)
```

```python
plt.plot(speed_list,mileage_list,label='$-\\frac{1}{28}s^2+3s-3$\nwhen 15  speed  70')
plt.axvline(0,color='black')
plt.axhline(0,color='black')
plt.xlabel('Speed (Miles per Hour)')
plt.ylabel('Gas Mileage')
plt.scatter(speed_list[mileage_list.index(max(mileage_list))],
            max(mileage_list),color='red',s=100,
            label=f'Optimal Gas Mileage is {max(mileage_list):.0f}\ngallons per mile when the car speed\nis 42 miles per hour')
plt.grid()
plt.legend(bbox_to_anchor=[0.55, 0.45]);
```
