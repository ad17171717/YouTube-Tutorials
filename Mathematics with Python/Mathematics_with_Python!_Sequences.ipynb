# **Sequences**

**A sequence is an ordered collection of objects. Sequences contain elements (also called members, or terms). Sequences can be infinite (going on forever) or finite. The same elements can appear multiple times at different positions in a sequence.**

<sup>[Sequence](https://mathworld.wolfram.com/Sequence.html) from Wolfram MathWorld by Eric Weisstein</sup>

<sup>[Sequence](https://en.wikipedia.org/wiki/Sequence) from Wikipedia.org</sup>

## **Arithmetic Sequence**

**An arithmetic sequence or arithmetic progression is a sequence of numbers such that the difference between the consecutive terms is constant.**

$a_n = a_{n-1} + d$

$\text{Where: $a$ is an element, $n$ is the index position and $d$ is the common difference}$

<sup>[Arithmetic Progression](https://en.wikipedia.org/wiki/Arithmetic_progression) from Wikipedia.org</sup>

$\text{Sequence 1 = $\{0,1,2,...,49\}$ where $\{a_n + 1\}_{n=0}^{49}$}$

```python
sequence_one = [num for num in range(50)]
print(sequence_one)
```

```python
for num_1,num_2 in zip(sequence_one[::2],sequence_one[1::2]):
  print(f'The difference between {num_2} and {num_1} is {num_2-num_1}')
```

$\text{Sequence 2 = $\{1,1\frac{1}{2},2,...,25\}$ where $\{a_n + \frac{1}{2}\}_{n=0}^{48}$}$

```python
sequence_two = [num/2 for num in range(2,51)]
print(sequence_two)
```

```python
for num_1,num_2 in zip(sequence_two[::2],sequence_two[1::2]):
  print(f'The difference between {num_2} and {num_1} is {num_2-num_1}')
```

```python
sequence_array = np.arange(start = 1, stop=25.5,step=.5)
print(type(sequence_array))
print(sequence_array)
```

## **Geometric Sequence**

**A geometric sequence, also known as a geometric progression, is a sequence of non-zero numbers where each term after the first is found by multiplying the previous one by a fixed, non-zero number called the common ratio.**

$a_n = ar^{n-1}$

$\text{Where: $a$ is an element, $n$ is the index position and $r$ is the common ratio}$

<sup>[Geometric progression](https://en.wikipedia.org/wiki/Geometric_progression) from Wikipedia.org</sup>

$\text{Sequence 3 = $\{3,9,27,...,14348907\}$ where $\{(3)3^{n}\}_{n=0}^{14}$}$

```python
geo_sequence = [3*3**(num) for num in range(15)]
print(geo_sequence)
```

```python
for num_1,num_2 in zip(geo_sequence[::2],geo_sequence[1::2]):
  print(f'{num_2} divided by {num_1} is {num_2/num_1}')
```

## **Fibonacci Sequence**

**The Fibonacci sequence is an integer sequence defined by a simple linear recurrence relation. The sequence appears in many settings in mathematics and in other sciences. In particular, the shape of many naturally occurring biological organisms is governed by the Fibonacci sequence and its close relative, the golden ratio.**

$F_n = F_{n-1} + F_{n - 2}$

$\text{Where: $F$ is a Fibonacci number and $n$ is the index position}$

<sup>[Fibonacci Sequence](https://brilliant.org/wiki/fibonacci-series/) from Brilliant.org</sup>

image.png

```python
fib_sequence = [0,1]

for iteration in range(15):
  fib_sequence.append(fib_sequence[-2] + fib_sequence[-1])

print(fib_sequence)
```

# **Real World Applications**

## **Geometric Sequence & Calculating the Value of an Interest Bearing Bank Deposit**

**A bank is offering an annual interest rate of 2.5% to its customers. If a customer deposits $1,000 with the bank how much money will they have in 5 years?**

$\text{Value of Deposit = Principal $\times$ (1+Interest Rate$)^{\text{Number of Periods}}$}$

```python
deposit_growth = [1000 * 1.025**(num) for num in range(6)]

for value,year in zip(deposit_growth,range(6)):
  print(f'The value of the deposit in year {year} is {value:,.2f}')
```

## **Fibonacci Sequence and Estimating Population Growth**

**In 1202, Leonardo Fibonacci investigated how fast rabbits could breed in ideal circumstances. In the scenario, a newly-born pair of rabbits, one male, one female, are put in a field. Rabbits are able to mate at the age of one month so that at the end of its second month a female can produce another pair of rabbits. Suppose that our rabbits never die and that the female always produces one new pair (one male, one female) every month from the second month on.**

**The Fibonacci Sequence is closely related to the logistic growth model, where the population growth is initially exponential, but as resources become scarce the population stays within the upper bound.**

<sup>[Fibonacci's Rabbits](https://r-knott.surrey.ac.uk/Fibonacci/fibnat.html#section1.1) from Ron Knott's web pages on Mathematics</sup>

<sup>[The Logistic Population Model](https://www2.clarku.edu/faculty/djoyce/ma121/logistic.pdf) by David Joyce</sup>

```python
rabbit_dict = {'Rabbit Population Estimate':fib_sequence,
               'Number of Months': [num for num in range(len(fib_sequence))]}
```

```python
fig = px.scatter(rabbit_dict, x='Number of Months', y='Rabbit Population Estimate', size = 'Rabbit Population Estimate',
                 text=['']*len(fib_sequence))

fig.update_layout(font_color = '#7030a0', 
        title={
        'text': 'Fibonnaci Sequence and Population Growth',
        'y':0.925,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.show()
```
