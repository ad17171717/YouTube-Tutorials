# **Stop Words**

**Stop words are frequently occurring, insignificant words that appear in text. For example, if we are conducting sentiment analysis on social media comments, stop words such as `this`, `is` and `the` do not assist in computing the sentiment of the comment.**

## **Removing Stop Words from a Simple String**

```python
test_str = 'the dog is hungry'
```

```python
stop_words = stopwords.words('english')
print(f'{type(stop_words)}\n{stop_words}')
```

```python
[word for word in test_str.split() if word not in stop_words]
```

## **Creating a Stop Word Removal Function**

```python
print(type(string.punctuation),'\n',string.punctuation)
```

```python
string.punctuation += ''
```

```python
def stop_word_removal(text, stop_word_corpus, punct_str):
    clean_text = ' '.join([word.lower() for word in text.split() if word.lower()
                 not in stop_word_corpus]).replace('\n',' ')
    return clean_text.translate(str.maketrans('', '', punct_str))
```

```python
stop_word_removal(poem,stop_words,string.punctuation)
```
