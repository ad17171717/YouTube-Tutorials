```python
from IPython.display import Image
import matplotlib.pyplot as plt
```

# **Rendering LaTeX Equations to a PNG Image**

```python
dot_product_equation_algebraic = r"$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i$"

fig, ax = plt.subplots(figsize=(6,3))
ax.text(0.5,0.5,dot_product_equation_algebraic,fontsize=50,ha='center',va='center')
ax.axis('off')

plt.savefig("dot_product_formula.png",
            dpi=1200,
            bbox_inches='tight',
            transparent=True)
plt.close()

Image(filename="dot_product_formula.png")
```

```python
sample_mean_equation = r"$\bar{x} = \frac{1}{n} \sum_{i=1}^n x_i$"

fig, ax = plt.subplots(figsize=(6,3))
ax.text(0.5,0.5,sample_mean_equation,fontsize=50,ha='center',va='center')
ax.axis('off')

plt.savefig("sample_mean_formula.png",
            dpi=1200,
            bbox_inches='tight',
            transparent=True)
plt.close()

Image(filename="sample_mean_formula.png")
```
