```python
import pandas as pd
import csv
```

```python
#generate the sample dataset
def create_example_csv(path: str = 'example.csv', n_rows: int = 100, n_cols: int = 50) -> None:
    '''
    Creates a CSV in the current working directory with columns id_1â€¦id_{n_cols} and n_rows rows
    where each cell in row i contains the integer i.
    '''

    #create column headers
    header = [f'id_{i}' for i in range(1, n_cols + 1)]
    with open(path, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        #write rows with a generator
        writer.writerows(
            ([i] * n_cols for i in range(n_rows))
        )

create_example_csv()
```

```python
#read in the sample dataset
df = pd.read_csv('example.csv')
```
