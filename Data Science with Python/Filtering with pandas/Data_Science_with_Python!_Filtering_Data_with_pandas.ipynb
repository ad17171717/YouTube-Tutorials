```python
import pandas as pd
```

```python
df = pd.read_csv('Meteorite_Landings.csv')
```

```python
df
```

# **Filtering Data**

## **Filter by Single Point Value**

```python
#double equals compares to see if one value is equal to another
df_single_point = df[df['year'] == 1880]
```

```python
df_single_point
```

## **Filter by in a Range of Values**

```python
df_year_range = df[df['year'] >= 1950]
```

```python
df_year_range.sort_values(by='year')
```

## **Filter in a Range Between Two Values**

```python
#be sure to use '&', not the 'and' operator
df_between = df[(df['year'] >= 1950) & (df['year'] <= 1999)]
```

```python
df_between.sort_values(by='year')
```

## **Filter Null Values Out**

```python
df.sort_values(by='GeoLocation')
```

```python
df_geo = df.dropna(subset=['GeoLocation']).sort_values(by='GeoLocation')
```

## **Filter Specified Value Out**

```python
df_geo.value_counts(subset='GeoLocation')
```

```python
print(df_geo.value_counts(subset='GeoLocation').index[0])
print(type(df_geo.value_counts(subset='GeoLocation').index[0]))
```

```python
df_geo_2 = df_geo[df_geo['GeoLocation'] != '(0.0, 0.0)']
```

```python
df_geo_2
```

```python
df_geo_2.value_counts(subset='GeoLocation')
```

## **Filter Columns using the `filter` Function**

```python
cols = ['name','mass (g)','year','GeoLocation']
```

```python
#axis 1 references columns, axis 0 references rows
df_cols = df.filter(items=cols,axis=1)
```

```python
df_cols
```

## **Filter Rows using the `filter` Function**

```python
#use list comprehension to create a list of rows to filter on
rows = [num for num in range(15)]
```

```python
df_rows = df.filter(items=rows,axis=0)
```

```python
df_rows
```

## **Filter Rows and Columns with the `loc` Function**

```python
#input a list of values to filter the rows first, then filter on columns as the next parameter
df_row_col = df.loc[rows,cols]
```

```python
df_row_col
```
